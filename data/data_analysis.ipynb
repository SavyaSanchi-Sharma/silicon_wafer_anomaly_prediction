```python
# Dataset 1 category name → unified class
d1_class_map = {
    "block etch": "open_etch",
    "coating bad": "coating_defect",
    "particle": "particle",
    "piq particle": "particle",
    "po contamination": "po_contamination",
    "scratch": "scratch",
    "sez burnt": "thermal_damage"
}

# dataset 2 category name → unified class
d2_class_map = {
    "bridge": "bridge",
    "particle": "particle",
    "scratch": "scratch"
}
```

```python
unified_classes = [
    "open_etch",
    "coating_defect",
    "particle",
    "po_contamination",
    "scratch",
    "bridge",
    "thermal_damage"
]
```

```python
def merge_coco_datasets(d1: dict, d2: dict) -> dict:
    unified_categories = []
    class_name_to_id = {}

    for idx, name in enumerate(UNIFIED_CLASSES):
        unified_categories.append({
            "id": idx,
            "name": name,
            "supercategory": "defect"
        })
        class_name_to_id[name] = idx

    def build_cat_id_map(categories, class_map):
        cat_id_to_unified = {}
        for c in categories:
            if c["name"] in class_map:
                unified_name = class_map[c["name"]]
                cat_id_to_unified[c["id"]] = class_name_to_id[unified_name]
        return cat_id_to_unified

    d1_cat_map = build_cat_id_map(d1["categories"], D1_CLASS_MAP)
    d2_cat_map = build_cat_id_map(d2["categories"], D2_CLASS_MAP)

    merged_images = []
    image_id_offset = 0

    d1_img_id_map = {}
    for img in d1["images"]:
        new_img = img.copy()
        new_img["id"] = image_id_offset
        d1_img_id_map[img["id"]] = image_id_offset
        merged_images.append(new_img)
        image_id_offset += 1

    d2_img_id_map = {}
    for img in d2["images"]:
        new_img = img.copy()
        new_img["id"] = image_id_offset
        d2_img_id_map[img["id"]] = image_id_offset
        merged_images.append(new_img)
        image_id_offset += 1

    merged_annotations = []
    ann_id = 0

    def add_annotations(annotations, img_map, cat_map):
        nonlocal ann_id
        for ann in annotations:
            if ann["category_id"] not in cat_map:
                continue

            new_ann = ann.copy()
            new_ann["id"] = ann_id
            new_ann["image_id"] = img_map[ann["image_id"]]
            new_ann["category_id"] = cat_map[ann["category_id"]]
            merged_annotations.append(new_ann)
            ann_id += 1

    add_annotations(d1["annotations"], d1_img_id_map, d1_cat_map)
    add_annotations(d2["annotations"], d2_img_id_map, d2_cat_map)

    merged_coco = {
        "info": {
            "year": "2024",
            "version": "merged_v1",
            "description": "Merged wafer defect datasets (D1 + D2)",
        },
        "licenses": d1.get("licenses", []),
        "categories": unified_categories,
        "images": merged_images,
        "annotations": merged_annotations
    }

    return merged_coco
```

```python
import os
import cv2
import json
import random
import uuid
from collections import defaultdict


def extract_clean_patches(
    coco: dict,
    image_dir: str,
    output_dir: str,
    patch_size: int = 128,
    margin: int = 16,
    max_patches_per_image: int = 3
):
    """
    Extract defect-free (clean) patches using bounding-box exclusion.

    Args:
        coco: merged COCO dict
        image_dir: path to 'train' images
        output_dir: path to save 'clean' images
        patch_size: size of square clean patch
        margin: safety margin around defect bboxes
        max_patches_per_image: limit patches per image
    """

    os.makedirs(output_dir, exist_ok=True)

    # Group annotations by image_id
    anns_by_image = defaultdict(list)
    for ann in coco["annotations"]:
        anns_by_image[ann["image_id"]].append(ann["bbox"])

    image_count = 0
    saved_count = 0

    for img_info in coco["images"]:
        img_id = img_info["id"]
        img_path = os.path.join(image_dir, img_info["file_name"])

        if not os.path.exists(img_path):
            continue

        image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if image is None:
            continue

        h, w = image.shape
        bboxes = anns_by_image.get(img_id, [])

        # Expand bboxes with margin
        forbidden = []
        for x, y, bw, bh in bboxes:
            x1 = max(0, int(x - margin))
            y1 = max(0, int(y - margin))
            x2 = min(w, int(x + bw + margin))
            y2 = min(h, int(y + bh + margin))
            forbidden.append((x1, y1, x2, y2))

        def is_clean(x1, y1, x2, y2):
            for fx1, fy1, fx2, fy2 in forbidden:
                if not (x2 <= fx1 or x1 >= fx2 or y2 <= fy1 or y1 >= fy2):
                    return False
            return True

        patches_found = 0
        attempts = 0
        max_attempts = 50

        while patches_found < max_patches_per_image and attempts < max_attempts:
            attempts += 1

            x1 = random.randint(0, w - patch_size)
            y1 = random.randint(0, h - patch_size)
            x2 = x1 + patch_size
            y2 = y1 + patch_size

            if not is_clean(x1, y1, x2, y2):
                continue

            patch = image[y1:y2, x1:x2]

            if patch.shape != (patch_size, patch_size):
                continue

            filename = f"clean_{img_id}_{uuid.uuid4().hex[:8]}.png"
            cv2.imwrite(os.path.join(output_dir, filename), patch)

            patches_found += 1
            saved_count += 1

        image_count += 1

    print(f"[INFO] Processed images: {image_count}")
    print(f"[INFO] Clean patches saved: {saved_count}")
```
